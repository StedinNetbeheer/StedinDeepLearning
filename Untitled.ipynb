## Introduction

Convolutional Neural Networks (ConvNets) are very similar to ordinary Neural Networks (NNets) as they are still made up of neurons which have learnable weights and biases. As in ordinary NNets, neurons receive from data or other layers, perform dot products and follows it with a non-linearity. The last (fully-connected) layer has a lost function depending on the type of problem we are tackling.

The difference between ordinary NNets and ConvNet architecture is that for ConvNets we make the explicit assumtion that model inputs are images. Certain properties are therefore implemented to make the forward function more efficient and reduce the number of parameters used.


 A regular 3-layer Neural Network.| A ConvNet arranges its neurons in three dimensions (width, height, depth).
- | - 
![neural_net2](neural_net2.jpeg) | ![cnn](cnn.jpeg)


_Reason for 3D layers_: In this notebook you will see that we use the hand-written digits from the MNIST dataset. These images are of size 28*28*3 (3 because of the 3 color channels). This means that a single fully connected neuron in the first hidden layer of a NNet would have 28*28*3 = 2,352 weights. It is easy to see that as the size of the images increases, this number would also increase. Also with more complex images we would also need a more complex network architecture, thus making our image classification problem more computationally expensive.

```python
# Import the relevant libraries
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
from keras.utils.np_utils import to_categorical
from keras.utils.vis_utils import plot_model

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import itertools

sns.set(style = 'white', context = 'notebook', palette = 'deep')
```

## Data prep

```python
# Download and split the data between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

The previous code loads the hand written digit data from the keras dataset and defines training and testing sets. Below we can see that we have allocated 60000 images for training and 10000 images for testing.

```python
# Input image dimensions
batch_size = 128
num_classes = 10

img_rows, img_cols = 28, 28
```

We must also make sure that, since this is a classification problem, we have the fairly similar occurences of each class in the training and testing datasets.

```python
# Plot the number of ocurrences for each number
sns.countplot(y_train)
```

### Label encoding

```python
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

### Data formatting

```python
if  K.image_data_format()=='channels_first':
    x_train = x_train.reshape(x_train.shape[0], 1, img_rows, img_cols)
    x_test = x_test.reshape(x_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
    x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

## Model building (99.49% accuracy)

We strongly suggest the use of the __keras__ package for model building. Keras is a high-level neural networks API, written in Python and capable of running on top of TensorFlow, CNTK, or Theano. Keras enables deep learning model building using both recurrent and convolutional nets and ability to run on both CPU and GPU. The following code is an example of how easily a ConvNet can be built in keras.

```python
# Build sequential model using keras
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile the above defined model
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 12, batch_size = batch_size, validation_split = 0.2, shuffle = True)

model.test_on_batch(x_test, y_test)
model.metrics_names
```

```python
print(history.history.keys())
#  "Accuracy"
plt.plot(history.history['acc'])
plt.style.use('ggplot')
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='lower right')
plt.show()
# "Loss"
plt.plot(history.history['loss'])
plt.style.use('ggplot')
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
score = model.evaluate(x_test, y_test, verbose = 0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
